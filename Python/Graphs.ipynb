# EXCEL

```python
def TO_EXCEL(data, size_end, excel_file, size_start = 1):
    '''
    TO_EXCEL(data, size_end, excel_file, size_start = 1)
    '''
    
    # Create a Pandas dataframe from the data.
    df = pd.DataFrame(data, index=data.index)

    # Create a Pandas Excel writer using XlsxWriter as the engine.
    #excel_file = 'graph.xlsx'
    sheet_name = 'Sheet1'

    writer = pd.ExcelWriter(excel_file, engine='xlsxwriter')
    df.to_excel(writer, sheet_name=sheet_name)

    # Access the XlsxWriter workbook and worksheet objects from the dataframe.
    workbook = writer.book
    worksheet = writer.sheets[sheet_name]

    # Create a chart object.
    chart = workbook.add_chart({'type': 'column'})

    # Configure the series of the chart from the dataframe data.
    for col_num in range(1, 4):
        chart.add_series({
            'name':       ['Sheet1', 0, col_num],
            'categories': ['Sheet1', 1, 0, size_end, 0],
            'values':     ['Sheet1', size_start, col_num, size_end, col_num],
            'fill':       {'color': brews['Set1'][col_num - 1]},
            'overlap':-10,
        })

    # Configure the chart axes.
    chart.set_x_axis({'name': 'Instancia'})
    chart.set_y_axis({'name': 'Tempo', 'major_gridlines': {'visible': False}})

    # Insert the chart into the worksheet.
    worksheet.insert_chart('H2', chart)

    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
```

```python
def nan_all_but_min(df):
    arr = df.values
    idx = np.argmin(arr, axis=1)
    newarr = np.full_like(arr, np.nan, dtype='float')
    newarr[np.arange(arr.shape[0]), idx] = arr[np.arange(arr.shape[0]), idx]
    df = pd.DataFrame(newarr, columns=df.columns, index=df.index)
    return df

df = pd.DataFrame(data)
print(df)

df = nan_all_but_min(df)
print(df)
```
